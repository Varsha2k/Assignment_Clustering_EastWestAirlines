```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

# Importing Dataset

```python
air=pd.read_csv("EastWestAirlines.csv")
air.head(6)
```

# EDA

```python
air.shape
```

```python
air.isna().sum()
```

```python
air.duplicated().sum()
```

```python
air1 = air.drop(labels = "ID#",axis=1 )
air1.head(5)
```

# Data Preparation

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
st_scaler = StandardScaler()
air_scaler = st_scaler.fit_transform(air1)
air_scaler = pd.DataFrame(data=air1,columns=['Balance','Qual_miles','cc1_miles','cc2_miles','cc3_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','Flight_trans_12','Days_since_enroll','Award?'])
air_scaler
```

```python
plt.figure(figsize=(20,12))
dendo = sch.dendrogram(sch.linkage(air_scaler,method='single'))
```

```python
plt.figure(figsize=(20,12))
dendo = sch.dendrogram(sch.linkage(air_scaler,method='complete'))
```

```python
plt.figure(figsize=(20,12))
dendo = sch.dendrogram(sch.linkage(air_scaler,method='average'))
```

```python
plt.figure(figsize=(20,12))
dendo = sch.dendrogram(sch.linkage(air_scaler,method='ward'))
```

# Model Building and Model Training

```python
hc = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc = hc.fit_predict(air_scaler)
y_hc
```

```python
clusters = pd.DataFrame(y_hc,columns=['cluster'])
clusters.head()
```

```python
air['clusters']= clusters
air.head(6)
```
